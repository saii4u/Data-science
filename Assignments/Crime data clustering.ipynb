HIERARICHAL CLUSTERING

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
from scipy import stats
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize, MinMaxScaler, StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
# Import Dataset

df = pd.read_csv(r"C:\Users\91739\Downloads\crime_data.csv")
df
```

```python
data_norm = pd.DataFrame(MinMaxScaler().fit_transform(df.iloc[:,1:]))
data_norm
```

```python
from scipy.cluster.hierarchy import linkage
z = linkage(data_norm, method='average', metric='euclidean')
plt.figure(figsize=(15,5))
plt.title("Hierarichal clustering dendrogram")
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()
```

#Insights from dendrograms

Here we can cut the dendrogram at a distance of 0.4 as we can separate the dendrogram into 4 clusters. Here, we can easily understand that all data points falls into seperate clusters. However, there is a mix in the blue cluster which we can further analyse through other clustering methods.

# exploring different types of linkages

```python
#complete linkage

from scipy.cluster.hierarchy import linkage
z = linkage(data_norm, method='complete', metric='euclidean')
plt.figure(figsize=(15,5))
plt.title("Hierarichal clustering dendrogram")
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()
```

```python
#centroid linkage

from scipy.cluster.hierarchy import linkage
z = linkage(data_norm, method='centroid', metric='euclidean')
plt.figure(figsize=(15,5))
plt.title("Hierarichal clustering dendrogram")
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()
```

```python
#single linkage

from scipy.cluster.hierarchy import linkage
z = linkage(data_norm, method='single', metric='euclidean')
plt.figure(figsize=(15,5))
plt.title("Hierarichal clustering dendrogram")
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()
```

```python
#ward linkage

from scipy.cluster.hierarchy import linkage
z = linkage(data_norm, method='ward', metric='euclidean')
plt.figure(figsize=(15,5))
plt.title("Hierarichal clustering dendrogram")
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()
```

# Exploring on Minkowski distance metric

```python
#Minkowski-complete linkage

from scipy.cluster.hierarchy import linkage
z = linkage(data_norm, method='complete', metric='minkowski')
plt.figure(figsize=(15,5))
plt.title("Hierarichal clustering dendrogram")
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()
```

```python
#Minkowski-single linkage

from scipy.cluster.hierarchy import linkage
z = linkage(data_norm, method='single', metric='minkowski')
plt.figure(figsize=(15,5))
plt.title("Hierarichal clustering dendrogram")
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()
```

```python
#Minkowski-average linkage

from scipy.cluster.hierarchy import linkage
z = linkage(data_norm, method='average', metric='minkowski')
plt.figure(figsize=(15,5))
plt.title("Hierarichal clustering dendrogram")
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()
```

**As we could that all the dendrograms are clear with the data points are majority falling into 4 clusters.**

```python
# Create Clusters 
h = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete').fit(data_norm)
cluster_labels = pd.Series(h.labels_)
df['cluster'] = cluster_labels
df
```

```python
# 5 clusters data
df.groupby(df.cluster).mean()
```

```python
# Plot Clusters according to Hierarchial clustering method
plt.figure(figsize=(15,5))  
plt.scatter(df['cluster'],df['Murder'], c=cluster_labels)

#conclusion
#Highest murders are happening in cluster "3"
```

K-MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python

crim1=pd.read_csv(r"C:\Users\91739\Downloads\crime_data.csv")
crim1
```

```python
from sklearn.preprocessing import StandardScaler
trans = StandardScaler()
crim1_norm = trans.fit_transform(crim1.iloc[:,1:])
crim1_norm
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(crim1_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(crim1_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crim1['Clust']=x
```

```python
crim1
```

```python
crim1.groupby(crim1.Clust).mean()
```

```python
# Plot Clusters according to Hierarchial clustering method
#checking highest number of Rapes happening in which cluster
plt.figure(figsize=(15,5))  
plt.scatter(crim1['Clust'],crim1['Rape'], c=cluster_labels)

#conclusion
#Highest murders are happening in cluster "0"
```

DBSCAN CLUSTERING FOR CRIME DATA

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crim3 = crim3.iloc[:,1:5]
crim3
```

```python
from sklearn.preprocessing import StandardScaler
crim3_norm = StandardScaler().fit_transform(crim3)
crim3_norm
```

```python
dbscan= DBSCAN(eps=2,min_samples=5)
dbscan.fit(crim3_norm)
```

```python
dbscan.labels_
```

```python
cl= pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
crim_db = pd.concat([crim3,cl],axis=1)
crim_db
```
